```python
#Method1
x = np.arange(1, 11)
y = 2*x+5

plt.title("Line Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x, y, 'oy')
plt.show()
```

```python
#Method2
x = np.arange(1, 11)
y = 2*x+5

plt.title("Line Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x, y, color = 'green', linestyle = 'dashed', linewidth = 3, 
        marker = 'o', markerfacecolor = 'yellow', markersize = 12)
plt.show()
```

```python
#line1
x1 = [1, 2, 3]
y1 = [2, 4, 1]

plt.plot(x1, y1, label = "Line1")

#Line2
x2 = [1, 2, 3]
y2 = [4, 1, 3]

plt.plot(x2, y2, label = "Line2")

plt.title("Two Lines on Same Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 1, 5, 2, 6]

plt.plot(x, y, color = 'cyan', linestyle = 'dashed', linewidth = 3, 
        marker = '<', markerfacecolor = 'orange', markersize = 12)

plt.xlim(1, 8)
plt.ylim(1, 8)

plt.title("Customization")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
#Compute x and y coordinates for sine and cosine graphs
x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#For Sine graph grid
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title("Sine Graph")

#For Cosine graph grid
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("Cos Graph")

plt.show()
```

# Scatter Plot

```python
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]

plt.scatter(x, y, color = 'b')
plt.show()
```

```python
x = [1, 1.5, 2, 2.5, 3, 3.5, 3.6]
y = [7.5, 8, 8.5, 9, 9.5, 10, 10.5]

x1 = [8, 8.5, 9, 9.5, 10, 10.5, 11]
y1 = [3, 3.5, 3.7, 4, 4.5, 5, 5.2]

plt.scatter(x, y, label = 'High Income Low Saving', color = 'r')
plt.scatter(x1, y1, label = 'Low Income High Saving', color = 'b')

plt.xlabel('Saving*100')
plt.ylabel('Income*100')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

# Pie Chart

```python
activities = ['eat', 'sleep', 'work', 'play']
slices = [3, 7, 8, 6]
colors = ['r', 'y', 'g', 'b']
plt.pie(slices, labels = activities, colors = colors, startangle = 90, 
       shadow = True, explode = (0, 0, 0.1, 0), radius = 1.5, 
        autopct = '%1.2f%%')

plt.show()
```

```python
#IPL Dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
match_data = pd.read_csv("matches.csv")
match_data
```

```python
# 8 most played venue in history of IPL from 2008 to 2017
```

```python
mpv = match_data['venue'].value_counts()
mpv = mpv[:8]

plt.bar(mpv.index, mpv.values, align = 'center')
plt.title("Most Played Venue")
plt.xlabel("Name of the Stadium")
plt.xticks(rotation = 90)
plt.ylabel("Count")
plt.show()
```

```python
sns.set_style('darkgrid')

mpv = match_data['venue'].value_counts()
mpv = mpv[:8]

plt.figure(figsize = (20, 6))

most_played = sns.barplot(mpv.index, mpv.values, alpha = 0.8)

plt.title("Most Played Venue", fontsize = 15)
plt.xlabel("Name of the Stadium", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
most_played.set_xticklabels(rotation = 90, labels = mpv.index, fontsize = 12)

plt.show()
```

```python
#top 10 Most player of the match
sns.set_style('darkgrid')

mpv = match_data['player_of_match'].value_counts()
mpv = mpv[:10]

plt.figure(figsize = (20, 6))

most_played = sns.barplot(mpv.index, mpv.values, alpha = 0.8, 
                          palette = 'winter')

plt.title("Most Played Players", fontsize = 15)
plt.xlabel("Name of the Player", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
most_played.set_xticklabels(rotation = 90, labels = mpv.index, fontsize = 12)

plt.show()
```

```python
#Matches played across each season

plt.figure(figsize = (20, 6))
sns.countplot(x = 'season', data = match_data)
plt.xlabel("Season", fontsize = 20)
plt.show()
```

```python
match_2017 = match_data[match_data['season']==2017]
df = match_2017[match_2017['toss_winner']==match_2017['winner']]
slices = [len(df), (59-len(df))]
labels = ['Yes', 'No']

plt.pie(slices, labels = labels, startangle = 90, shadow = True, 
       explode = (0, 0.1), colors = ['grey', 'cyan'], autopct = '%1.2f%%')
plt.show()
```
